```python
#Importing Libraries.
from datetime import date
import pywhatkit
from tkinter import *
import tkinter as tk
import tkinter.messagebox
import random
import time
import os


root = Tk()
time1 = ''
root.title("AUTOMATE WHATSAPP")
root.geometry("1920x1080+0+0")
root.config(bg='#F0F8FF')

#Creating the Frames
Tops = Frame(root, width=1600, height=10, bg="#F0F8FF",relief=SUNKEN)
Tops.pack(side=TOP)
f1 = Frame(root, width=1080, height=700, bg="#F0F8FF",relief=SUNKEN)
f1.pack(side=LEFT)
f2 = Frame(root, width=500, height=700, bg="#F0F8FF",relief=SUNKEN)
f2.pack(side=BOTTOM)

#Creating dynamic clock
def tick():
    global time1
    time2 = time.strftime("%Y-%m-%d %H:%M:%S")
    if time2 != time1:
        time1 = time2
        clock.config(text=time2)
    clock.after(200, tick)

#Header
heading = Label(Tops,font=("Century",70,"bold","underline"),text="AUTOMATE WHATSAPP",fg='Black',bg="#F0F8FF",bd=10, anchor='w')
heading.grid(row=0,column=0)
clock = Label(Tops,font=("Century",35,"bold","underline"),text=tick,fg='Black',bg="#F0F8FF",bd=10, anchor='w')
clock.grid(row=1,column=0)

tick()

#Defining clear button
def clrfuncno():
    enternumber.set("")
enternumber = StringVar()

def clrfuncmsg():
    entermessage.set("")
entermessage = StringVar()


time_hour = StringVar()
time_min = StringVar()

#Defining Send Button
def send():
    number=enternumber.get()
    number = '+91' + number
    message=entermessage.get()
    hour=int(time_hour.get())
    min=int(time_min.get())

    pywhatkit.sendwhatmsg(number,message,hour,min)

#Defining Reset button
def resetall():
    everything.set("")
    entermessage.set("")
    enternumber.set("")
    time_hour.set("")
    time_min.set("")

everything = StringVar()

#Defining Error Button


#Blanks
lblblank1 = Label(f1,font=("Century",30,"bold"),text="Pa:",fg='#F0F8FF',bg="#F0F8FF",bd=20, anchor='w')
lblblank1.grid(row=1,column=0)
lblblank2 = Label(f1,font=("Century",30,"bold"),text="Pasdaf:",fg='#F0F8FF',bg="#F0F8FF",bd=20, anchor='w')
lblblank2.grid(row=4,column=3)

#Creating the GUI

lblphonenumber = Label(f1,font=("Century",30,"bold"),text="Phone Number:",fg='Black',bg="#F0F8FF",bd=30, anchor='w')
lblphonenumber.grid(row=1,column=1)
txtphonenumber = Entry(f1,font=("Century",10,"bold"),textvariable=enternumber,bd=10,insertwidth=4,bg="white",justify="center")
txtphonenumber.grid(row=1,column=2)

lbltime = Label(f1,font=("Century",30,"bold"),text="Time:",fg='Black',bg="#F0F8FF",bd=30, anchor='w')
lbltime.grid(row=2,column=0)
lbltime = Label(f1,font=("Century",20,"bold"),text="(24-hr):",fg='Black',bg="#F0F8FF",bd=30, anchor='w')
lbltime.grid(row=3,column=0)

lblhour = Label(f1,font=("Century",20,"bold"),text="Hour",fg='Black',bg="#F0F8FF",bd=30, anchor='w')
lblhour.grid(row=3,column=1)
txthour = Entry(f1,font=("Century",10,"bold"),textvariable=time_hour,bd=10,insertwidth=4,bg="white",justify="center")
txthour.grid(row=2,column=1)

lblminutes = Label(f1,font=("Century",20,"bold"),text="Minute",fg='Black',bg="#F0F8FF",bd=30, anchor='w')
lblminutes.grid(row=3,column=2)
txtminutes = Entry(f1,font=("Century",10,"bold"),textvariable=time_min,bd=10,insertwidth=4,bg="white",justify="center")
txtminutes.grid(row=2,column=2)

lblmessage = Label(f1,font=("Century",30,"bold"),text="Message",fg='Black',bg="#F0F8FF",bd=30, anchor='w')
lblmessage.grid(row=1,column=4)
txtmessage = Entry(f1,font=("Century",20),textvariable=entermessage,bd=10,insertwidth=8,bg="white",justify="left")
txtmessage.grid(row=1,column=5,padx=20)


#Button
btnreset = Button(f1,padx=16,pady=8,bd=8,fg='black',font=("Century",20,"bold"),text='RESET',command=resetall,bg='powder blue').grid(row=5,column=2)
btnclrno = tk.Button(f1,padx=4,pady=2,bd=8,fg='black',font=("Century",14,"bold"),text='CLEAR',command=clrfuncno,bg='powder blue').grid(row=1,column=3)
btnclrmsg = Button(f1,padx=4,pady=2,bd=8,fg='black',font=("Century",14,"bold"),text='CLEAR',command=clrfuncmsg,bg='powder blue').grid(row=1,column=6)
btnsend = Button(f1,padx=16,pady=8,bd=8,fg='black',font=("Century",20,"bold"),text='SEND',command=send,bg='powder blue',).grid(row=5,column=3)
btnerror = Button(f1,padx=16,pady=8,bd=8,fg='black',font=("Century",20,"bold"),text='Error Check',bg='red',).grid(row=5,column=4)

root.mainloop()
```
